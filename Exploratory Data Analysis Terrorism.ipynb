```python

data.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country',
                     'region_txt':'Region','provstate':'Province','city':'City','latitude':'Latitude',
                     'longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype',
                     'targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Killed',
                     'nwound':'Wounded'},inplace = True)
```

```python

data = data[['Year','Month','Day','Country','Province','Region','City','Latitude','Longitude',"Attacktype",
             'Killed','Wounded','target1','Summary','Group','Targettype','Weapon','motive']]
data.head()
```

Step 3 : Visualization



```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot= True)
plt.show()
```

```python
#Top 10 Weapons used
plt.subplots(figsize=(20,10))
sns.barplot(data['Weapon'].value_counts()[:10].index, data['Weapon'].value_counts()[:10].values, palette='colorblind')
plt.title("Top 10 Weapons Used ", fontsize=24)
plt.xlabel('Weapon Type', fontsize=18)
plt.ylabel('No. Of Weapons', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 Countries With Highest Terror Attacks
plt.subplots(figsize=(20,10))
country = data['Country'].unique()
countries_count = data['Country'].value_counts(dropna = False).sort_index()
sns.barplot(x=country, y=countries_count)
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='Dark2')
plt.title("Top 10 Countries With Highest Terror Attacks", fontsize=24)
plt.xlabel('Countries', fontsize=18)
plt.ylabel('Attacks', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 Cities With Highest Terror Attacks
plt.subplots(figsize=(20,10))
sns.barplot(data['City'].value_counts()[:10].index,data['City'].value_counts()[:10].values,palette= 'nipy_spectral')
plt.title("Top 10 Cities With Highest Terror Attacks", fontsize=24)
plt.xlabel('Cities', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Terrorist Activities By Region In Each Year
plt.subplots(figsize=(12,7))
sns.kdeplot(data['Year'], hue=data['Region'])
plt.title('Terrorist Activities By Region', fontsize=24)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Frequency Of Attacks', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Deaths in Terror Attacks per Year
plt.subplots(figsize=(20,10))
data.groupby(['Year'])['Killed'].sum().sort_values(ascending = False).head(20).plot(kind = 'pie')
plt.title("Total Deaths in Year", fontsize=24)
plt.xlabel('Attack Year', fontsize=18)
plt.ylabel('No. Of Deaths', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 Terrorist Organisation with Highest Terror Attacks
plt.figure(figsize = (12,7))
sns.barplot(x = data['Group'].value_counts()[1:11].values, y = data['Group'].value_counts()[1:11].index,palette='magma')
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks', fontsize=24)
plt.ylabel('Terror Group Name', fontsize=18)
plt.xlabel('Number of Attacks', fontsize=18)
plt.xticks(rotation= 90)
plt.show()
```
